```python
import os
import json
from concurrent.futures import ThreadPoolExecutor

# JSON 파일이 저장된 최상위 경로 설정
base_folder_path = '/content/drive/MyDrive/dataset'

# 모든 JSON 파일 경로 재귀적으로 가져오기
json_files = []
for root, dirs, files in os.walk(base_folder_path):
    for file in files:
        if file.endswith('.json'):
            json_files.append(os.path.join(root, file))

print(f'Total JSON files found: {len(json_files)}')

# JSON 파일 로드 함수
def load_json(file):
    with open(file, 'r', encoding='utf-8') as f:
        return json.load(f)

# 병렬로 JSON 파일 로드
annotations = []
with ThreadPoolExecutor() as executor:
    results = executor.map(load_json, json_files)
    for data in results:
        for scene in data['scene']['data']:
            for occupant in scene['occupant']:
                annotation = {
                    'img_name': scene['img_name'],
                    'body_b_box': occupant['body_b_box'],
                    'face_b_box': occupant['face_b_box'],
                    'action': occupant['action'],
                    'emotion': occupant['emotion']
                }
                annotations.append(annotation)

print(f'Total annotations loaded: {len(annotations)}')
print(annotations[:10])  # 처음 10개의 항목을 출력하여 확인
```
